# ORDINAL ENCODER

```python
#ordinal encoding for column B income
df=pd.read_csv('customers.csv')
```

```python
df
```

```python
df.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df.income.unique()
```

```python
income_range=['high','medium','low']
```

```python
ordinal=OrdinalEncoder(categories=[income_range])
```

```python
encoded_values=ordinal.fit_transform(df[['income']])
```

```python
new_income=pd.DataFrame(encoded_values,columns=['income'])
```

```python
new_income
```

```python
df.income=new_income
```

```python
df
```

```python
df.head()
```
